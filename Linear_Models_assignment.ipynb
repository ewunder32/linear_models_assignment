**Q1.** Let's explore multiple linear regression in a two-variable case, to build more intuition about what is happening.

Suppose the model is
$$
\hat{y}_i = b_0 + b_1 z_{i1} + b_2 z_{i2}
$$
Assume that $z_{ij}$ is centered or de-meaned, so that $z_{ij} = x_{ij} - m_j$ where $m_j$ is the mean of variable $j$ and $x_{ij}$ is the original value of variable $j$ for observation $i$. Notice that this implies
$$
\dfrac{1}{N} \sum_{i=1}^N z_{ij} = 0
$$
which will simplify your calculations below substantially!

1. Write down the SSE for this model.
2. Take partial derivatives with respect to $b_0$, $b_1$, and $b_2$.
3. Verify that the average error is zero and $e \cdot z =0$ at the optimum, just as in the single linear regression case.
4. Show that the optimal intercept is $b_0^* = \bar{y}$. Eliminate $b_0^*$ from the remaining equations, and focus on $b_1$ and $b_2$.
5. Write your results as a matrix equation in the form "$Ab=C$". These are called the **normal equations**.
6. Divide both sides by $N$ and substitute $z_{ij} = x_{ij} - m_j$ back into your normal equations for $x_{ij}$. What is the matrix $A$? What is the vector $C$? Explain the intuition of your discovery.



1. **Write down the SSE for this model.**

The Sum of Squared Errors (SSE) is given by:
$$
SSE = \sum_{i=1}^N (y_i - \hat{y}_i)^2 = \sum_{i=1}^N (y_i - (b_0 + b_1 z_{i1} + b_2 z_{i2}))^2
$$

2. **Take partial derivatives with respect to $b_0$, $b_1$, and $b_2$.**

To find the values of $b_0, b_1, b_2$ that minimize the SSE, i take the partial derivatives with respect to each parameter and set them equal to zero

Partial derivative with respect to $b_0$:
$$
\dfrac{\partial SSE}{\partial b_0} = \sum_{i=1}^N 2(y_i - b_0 - b_1 z_{i1} - b_2 z_{i2})(-1) = -2 \sum_{i=1}^N (y_i - b_0 - b_1 z_{i1} - b_2 z_{i2})
$$
Setting this to zero:
$$
\sum_{i=1}^N (y_i - b_0 - b_1 z_{i1} - b_2 z_{i2}) = 0
$$
$$
\sum_{i=1}^N y_i - \sum_{i=1}^N b_0 - \sum_{i=1}^N b_1 z_{i1} - \sum_{i=1}^N b_2 z_{i2} = 0
$$
Since $\sum_{i=1}^N z_{ij} = 0$ for centered variables:
$$
\sum_{i=1}^N y_i - N b_0 - b_1 \sum_{i=1}^N z_{i1} - b_2 \sum_{i=1}^N z_{i2} = 0
$$
$$
\sum_{i=1}^N y_i - N b_0 = 0
$$

Partial derivative with respect to $b_1$:
$$
\dfrac{\partial SSE}{\partial b_1} = \sum_{i=1}^N 2(y_i - b_0 - b_1 z_{i1} - b_2 z_{i2})(-z_{i1}) = -2 \sum_{i=1}^N z_{i1}(y_i - b_0 - b_1 z_{i1} - b_2 z_{i2})
$$
Setting this to zero:
$$
\sum_{i=1}^N z_{i1}(y_i - b_0 - b_1 z_{i1} - b_2 z_{i2}) = 0
$$
$$
\sum_{i=1}^N z_{i1} y_i - \sum_{i=1}^N b_0 z_{i1} - \sum_{i=1}^N b_1 z_{i1}^2 - \sum_{i=1}^N b_2 z_{i1} z_{i2} = 0
$$
Since $\sum_{i=1}^N z_{i1} = 0$:
$$
\sum_{i=1}^N z_{i1} y_i - b_0 \sum_{i=1}^N z_{i1} - b_1 \sum_{i=1}^N z_{i1}^2 - b_2 \sum_{i=1}^N z_{i1} z_{i2} = 0
$$
$$
\sum_{i=1}^N z_{i1} y_i - b_1 \sum_{i=1}^N z_{i1}^2 - b_2 \sum_{i=1}^N z_{i1} z_{i2} = 0
$$

Partial derivative with respect to $b_2$:
$$
\dfrac{\partial SSE}{\partial b_2} = \sum_{i=1}^N 2(y_i - b_0 - b_1 z_{i1} - b_2 z_{i2})(-z_{i2}) = -2 \sum_{i=1}^N z_{i2}(y_i - b_0 - b_1 z_{i1} - b_2 z_{i2})
$$
Setting this to zero:
$$
\sum_{i=1}^N z_{i2}(y_i - b_0 - b_1 z_{i1} - b_2 z_{i2}) = 0
$$
$$
\sum_{i=1}^N z_{i2} y_i - \sum_{i=1}^N b_0 z_{i2} - \sum_{i=1}^N b_1 z_{i1} z_{i2} - \sum_{i=1}^N b_2 z_{i2}^2 = 0
$$
Since $\sum_{i=1}^N z_{i2} = 0$:
$$
\sum_{i=1}^N z_{i2} y_i - b_0 \sum_{i=1}^N z_{i2} - b_1 \sum_{i=1}^N z_{i1} z_{i2} - b_2 \sum_{i=1}^N z_{i2}^2 = 0
$$
$$
\sum_{i=1}^N z_{i2} y_i - b_1 \sum_{i=1}^N z_{i1} z_{i2} - b_2 \sum_{i=1}^N z_{i2}^2 = 0
$$

3. **Verify that the average error is zero and $e \cdot z =0$ at the optimum, just as in the single linear regression case.**

From the partial derivative with respect to $b_0$:
$$
\sum_{i=1}^N (y_i - b_0 - b_1 z_{i1} - b_2 z_{i2}) = 0
$$
 $e_i = y_i - \hat{y}_i$. So, $\sum_{i=1}^N e_i = 0$

  $\frac{1}{N}\sum_{i=1}^N e_i = 0$.

From the partial derivative with respect to $b_1$:
$$
\sum_{i=1}^N z_{i1}(y_i - b_0 - b_1 z_{i1} - b_2 z_{i2}) = 0
$$

$\sum_{i=1}^N z_{i1} e_i = 0$

the first centered predictor vector being zero: $e \cdot z_1 = 0$.

From the partial derivative with respect to $b_2$:
$$
\sum_{i=1}^N z_{i2}(y_i - b_0 - b_1 z_{i1} - b_2 z_{i2}) = 0
$$
This can be written as $\sum_{i=1}^N z_{i2} e_i = 0$, which is the dot product of the error vector and the second centered predictor vector being zero: $e \cdot z_2 = 0$.

having the average error as zero and error as orthogonal to each predictor are first-order conditions for minimizing the SSE and are analogous to the single linear regression case

4. **Show that the optimal intercept is $b_0^* = \bar{y}$. Eliminate $b_0^*$ from the remaining equations, and focus on $b_1$ and $b_2$.**

From the first normal equation (derived from the partial derivative with respect to $b_0$):
$$
\sum_{i=1}^N y_i - N b_0 = 0
$$
$$
N b_0 = \sum_{i=1}^N y_i
$$
$$
b_0 = \dfrac{1}{N} \sum_{i=1}^N y_i = \bar{y}
$$
the optimal intercept $b_0^*$ is indeed the mean of the dependent variable $\bar{y}$.

next eliminate $b_0$ from the other two normal equations. Since we are at the optimum, we substitute $b_0^* = \bar{y}$. Also, since $\sum_{i=1}^N z_{ij} = 0$, we have $\sum_{i=1}^N b_0^* z_{ij} = b_0^* \sum_{i=1}^N z_{ij} = \bar{y} \cdot 0 = 0$.

The second normal equation:
$$
\sum_{i=1}^N z_{i1} y_i - b_0 \sum_{i=1}^N z_{i1} - b_1 \sum_{i=1}^N z_{i1}^2 - b_2 \sum_{i=1}^N z_{i1} z_{i2} = 0
$$
Substituting $b_0^* = \bar{y}$ and using $\sum_{i=1}^N z_{i1} = 0$:
$$
\sum_{i=1}^N z_{i1} y_i - \bar{y} \cdot 0 - b_1 \sum_{i=1}^N z_{i1}^2 - b_2 \sum_{i=1}^N z_{i1} z_{i2} = 0
$$
$$
\sum_{i=1}^N z_{i1} y_i = b_1 \sum_{i=1}^N z_{i1}^2 + b_2 \sum_{i=1}^N z_{i1} z_{i2}
$$

The third normal equation:
$$
\sum_{i=1}^N z_{i2} y_i - b_0 \sum_{i=1}^N z_{i2} - b_1 \sum_{i=1}^N z_{i1} z_{i2} - b_2 \sum_{i=1}^N z_{i2}^2 = 0
$$
Substituting $b_0^* = \bar{y}$ and using $\sum_{i=1}^N z_{i2} = 0$:
$$
\sum_{i=1}^N z_{i2} y_i - \bar{y} \cdot 0 - b_1 \sum_{i=1}^N z_{i1} z_{i2} - b_2 \sum_{i=1}^N z_{i2}^2 = 0
$$
$$
\sum_{i=1}^N z_{i2} y_i = b_1 \sum_{i=1}^N z_{i1} z_{i2} + b_2 \sum_{i=1}^N z_{i2}^2
$$

So the two equations focusing on $b_1$ and $b_2$ are:
$$
\sum_{i=1}^N z_{i1} y_i = b_1 \sum_{i=1}^N z_{i1}^2 + b_2 \sum_{i=1}^N z_{i1} z_{i2}
$$
$$
\sum_{i=1}^N z_{i2} y_i = b_1 \sum_{i=1}^N z_{i1} z_{i2} + b_2 \sum_{i=1}^N z_{i2}^2
$$

5. **Write your results as a matrix equation in the form "$Ab=C$". These are called the normal equations.**

 write the system of equations for $b_1$ and $b_2$ in matrix form. Let $b = \begin{pmatrix} b_1 \\ b_2 \end{pmatrix}$.

The equations are:
$$
\left(\sum_{i=1}^N z_{i1}^2\right) b_1 + \left(\sum_{i=1}^N z_{i1} z_{i2}\right) b_2 = \sum_{i=1}^N z_{i1} y_i
$$
$$
\left(\sum_{i=1}^N z_{i1} z_{i2}\right) b_1 + \left(\sum_{i=1}^N z_{i2}^2\right) b_2 = \sum_{i=1}^N z_{i2} y_i
$$

In matrix form $Ab=C$:
$$
\begin{pmatrix}
\sum_{i=1}^N z_{i1}^2 & \sum_{i=1}^N z_{i1} z_{i2} \\
\sum_{i=1}^N z_{i1} z_{i2} & \sum_{i=1}^N z_{i2}^2
\end{pmatrix}
\begin{pmatrix}
b_1 \\
b_2
\end{pmatrix}
=
\begin{pmatrix}
\sum_{i=1}^N z_{i1} y_i \\
\sum_{i=1}^N z_{i2} y_i
\end{pmatrix}
$$

Here, $A = \begin{pmatrix}
\sum_{i=1}^N z_{i1}^2 & \sum_{i=1}^N z_{i1} z_{i2} \\
\sum_{i=1}^N z_{i1} z_{i2} & \sum_{i=1}^N z_{i2}^2
\end{pmatrix}$, $b = \begin{pmatrix} b_1 \\ b_2 \end{pmatrix}$, and $C = \begin{pmatrix}
\sum_{i=1}^N z_{i1} y_i \\
\sum_{i=1}^N z_{i2} y_i
\end{pmatrix}$.

6. **Divide both sides by $N$ and substitute $z_{ij} = x_{ij} - m_j$ back into your normal equations for $x_{ij}$. What is the matrix $A$? What is the vector $C$? Explain the intuition of your discovery.**

Dividing the matrix equation by $N$:
$$
\dfrac{1}{N}
\begin{pmatrix}
\sum_{i=1}^N z_{i1}^2 & \sum_{i=1}^N z_{i1} z_{i2} \\
\sum_{i=1}^N z_{i1} z_{i2} & \sum_{i=1}^N z_{i2}^2
\end{pmatrix}
\begin{pmatrix}
b_1 \\
b_2
\end{pmatrix}
=
\dfrac{1}{N}
\begin{pmatrix}
\sum_{i=1}^N z_{i1} y_i \\
\sum_{i=1}^N z_{i2} y_i
\end{pmatrix}
$$
$$
\begin{pmatrix}
\dfrac{1}{N}\sum_{i=1}^N z_{i1}^2 & \dfrac{1}{N}\sum_{i=1}^N z_{i1} z_{i2} \\
\dfrac{1}{N}\sum_{i=1}^N z_{i1} z_{i2} & \dfrac{1}{N}\sum_{i=1}^N z_{i2}^2
\end{pmatrix}
\begin{pmatrix}
b_1 \\
b_2
\end{pmatrix}
=
\begin{pmatrix}
\dfrac{1}{N}\sum_{i=1}^N z_{i1} y_i \\
\dfrac{1}{N}\sum_{i=1}^N z_{i2} y_i
\end{pmatrix}
$$

Now, substitute $z_{ij} = x_{ij} - m_j$. We know that $\dfrac{1}{N}\sum_{i=1}^N (x_{ij} - m_j) = \bar{x}_j - m_j$. Since $m_j$ is the mean of $x_{j}$, $m_j = \bar{x}_j$, so $\dfrac{1}{N}\sum_{i=1}^N z_{ij} = 0$.

The terms in the matrix $A$ are sample variances and covariances of the centered variables:
$\dfrac{1}{N}\sum_{i=1}^N z_{i1}^2 = \dfrac{1}{N}\sum_{i=1}^N (x_{i1} - \bar{x}_1)^2 = \text{Var}(x_1)$
$\dfrac{1}{N}\sum_{i=1}^N z_{i2}^2 = \dfrac{1}{N}\sum_{i=1}^N (x_{i2} - \bar{x}_2)^2 = \text{Var}(x_2)$
$\dfrac{1}{N}\sum_{i=1}^N z_{i1} z_{i2} = \dfrac{1}{N}\sum_{i=1}^N (x_{i1} - \bar{x}_1)(x_{i2} - \bar{x}_2) = \text{Cov}(x_1, x_2)$

So, the matrix $A$ is the sample covariance matrix of the predictors $z_1$ and $z_2$ (which is the same as the covariance matrix of $x_1$ and $x_2$ when they are centered):
$$
A = \begin{pmatrix}
\text{Var}(z_1) & \text{Cov}(z_1, z_2) \\
\text{Cov}(z_1, z_2) & \text{Var}(z_2)
\end{pmatrix} = \begin{pmatrix}
\text{Var}(x_1) & \text{Cov}(x_1, x_2) \\
\text{Cov}(x_1, x_2) & \text{Var}(x_2)
\end{pmatrix}
$$

The terms in the vector $C$ are the sample covariances between the centered predictors and the dependent variable $y$:
$\dfrac{1}{N}\sum_{i=1}^N z_{i1} y_i = \dfrac{1}{N}\sum_{i=1}^N (x_{i1} - \bar{x}_1)(y_i - \bar{y} + \bar{y}) = \dfrac{1}{N}\sum_{i=1}^N (x_{i1} - \bar{x}_1)(y_i - \bar{y}) + \dfrac{1}{N}\sum_{i=1}^N (x_{i1} - \bar{x}_1)\bar{y} = \text{Cov}(x_1, y) + \bar{y} \dfrac{1}{N}\sum_{i=1}^N (x_{i1} - \bar{x}_1) = \text{Cov}(x_1, y) + \bar{y} \cdot 0 = \text{Cov}(x_1, y)$.

Similarly, $\dfrac{1}{N}\sum_{i=1}^N z_{i2} y_i = \text{Cov}(x_2, y)$.

So, the vector $C$ is the sample covariance vector between the predictors and the dependent variable:
$$
C = \begin{pmatrix}
\text{Cov}(z_1, y) \\
\text{Cov}(z_2, y)
\end{pmatrix} = \begin{pmatrix}
\text{Cov}(x_1, y) \\
\text{Cov}(x_2, y)
\end{pmatrix}
$$

The normal equations in terms of variances and covariances are:
$$
\begin{pmatrix}
\text{Var}(x_1) & \text{Cov}(x_1, x_2) \\
\text{Cov}(x_1, x_2) & \text{Var}(x_2)
\end{pmatrix}
\begin{pmatrix}
b_1 \\
b_2
\end{pmatrix}
=
\begin{pmatrix}
\text{Cov}(x_1, y) \\
\text{Cov}(x_2, y)
\end{pmatrix}
$$

**Intuition of my discovery:**

The regression coefficients ($b_1$ and $b_2$) are determined by the variances and covariances of the predictors and their covariances with the dependent variable. The matrix $A$ shows the relationships between the predictors themselves, while the vector $C$ shows the relationships between each predictor and the outcome. Essentially, the more a predictor varies and the more it covaries with the outcome, the larger its corresponding coefficient will tend to be. The matrix $A$ basically "standardizes" the relationships, and this allows us to solve for the unique contribution of each predictor in relation to the variance in $y$.

```python
#Q2.1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/cars_hw.csv')

# Display the first few rows and information
display(df.head())
df.info()
```

```python
#Q2.1
# Examine descriptive statistics of numerical columns
display(df.describe())

# Plot histograms for numerical columns to visualize distributions
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
df[numerical_cols].hist(figsize=(10, 8))
plt.tight_layout()
plt.show()

# Plot boxplots for numerical columns to visualize outliers
df[numerical_cols].plot(kind='box', subplots=True, layout=(3, 2), figsize=(10, 10))
plt.tight_layout()
plt.show()
```

```python
#Q2.1
# Drop the 'Unnamed: 0' column as it is just an index
df = df.drop('Unnamed: 0', axis=1)

# Apply log transformation to 'Price' and 'Mileage_Run' to handle skewness and outliers
df['Price_log'] = np.log1p(df['Price']) # Using log1p to handle potential zero values if any, although min price is > 0
df['Mileage_Run_log'] = np.log1p(df['Mileage_Run'])

# Examine the distributions of the transformed variables
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(df['Price_log'], kde=True, ax=axes[0])
axes[0].set_title('Distribution of Log-Transformed Price')
sns.histplot(df['Mileage_Run_log'], kde=True, ax=axes[1])
axes[1].set_title('Distribution of Log-Transformed Mileage_Run')
plt.tight_layout()
plt.show()

# Check for outliers in numerical columns after transformation using box plots
numerical_cols_transformed = ['Make_Year', 'Price_log', 'Mileage_Run_log']
df[numerical_cols_transformed].plot(kind='box', subplots=True, layout=(1, 3), figsize=(15, 5))
plt.tight_layout()
plt.show()
```

```python
#Q2.2
# Summarize the original 'Price' variable
print("Summary of original 'Price' variable:")
display(df['Price'].describe())

# Create a kernel density plot for the original 'Price'
plt.figure(figsize=(8, 5))
sns.kdeplot(df['Price'], fill=True)
plt.title('Kernel Density Plot of Car Prices')
plt.xlabel('Price')
plt.ylabel('Density')
plt.show()

# Summarize 'Price' by 'Make'
print("\nSummary of 'Price' grouped by 'Make':")
display(df.groupby('Make')['Price'].describe())

# Create a grouped kernel density plot for 'Price' by 'Make'
plt.figure(figsize=(12, 8))
sns.kdeplot(data=df, x='Price', hue='Make', common_norm=False)
plt.title('Kernel Density Plot of Car Prices by Make')
plt.xlabel('Price')
plt.ylabel('Density')
plt.show()
```

```python
#Q2.5

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Define a list of polynomial degrees to iterate through
degrees = range(1, 6) # Start from degree 1 (linear) up to 5

# Create empty lists to store R-squared and RMSE values for the test set
r2_scores_test = []
rmse_scores_test = []

# Loop through the defined polynomial degrees
for degree in degrees:
    print(f"\nEvaluating model with polynomial degree: {degree}")
    # Instantiate PolynomialFeatures with the current degree
    polynomial_converter = PolynomialFeatures(degree=degree, include_bias=False)

    # Fit and transform the numerical training data
    X_train_poly = polynomial_converter.fit_transform(X_train_numeric)

    # Transform the numerical testing data
    X_test_poly = polynomial_converter.transform(X_test_numeric)

    # Instantiate a LinearRegression model
    model_poly = LinearRegression()

    # Train the linear regression model on the polynomial-transformed training data
    model_poly.fit(X_train_poly, y_train)

    # Predict the target variable for the polynomial-transformed testing data
    y_test_pred_poly = model_poly.predict(X_test_poly)

    # Calculate the R-squared and RMSE for the test set predictions
    r2_test_poly = r2_score(y_test, y_test_pred_poly)
    rmse_test_poly = np.sqrt(mean_squared_error(y_test, y_test_pred_poly))

    # Append the calculated R-squared and RMSE to their respective lists
    r2_scores_test.append(r2_test_poly)
    rmse_scores_test.append(rmse_test_poly)

    print(f"Test R^2: {r2_test_poly:.4f}")
    print(f"Test RMSE: {rmse_test_poly:.4f}")

    # Check if the R-squared value is negative
    if r2_test_poly < 0:
        print(f"R^2 went negative at degree {degree}. Stopping.")
        break

# After the loop, identify the degree that yielded the highest R-squared on the test set before R-squared turned negative.
# We need to consider the degrees for which R^2 was not negative
valid_degrees = [d for i, d in enumerate(degrees) if i < len(r2_scores_test) and r2_scores_test[i] >= 0]
valid_r2_scores = [r2 for r2 in r2_scores_test if r2 >= 0]

if valid_r2_scores:
    best_degree_index = np.argmax(valid_r2_scores)
    best_polynomial_degree = valid_degrees[best_degree_index]
    best_poly_r2 = valid_r2_scores[best_degree_index]
    best_poly_rmse = rmse_scores_test[best_degree_index] # Get corresponding RMSE
else:
    best_polynomial_degree = None
    best_poly_r2 = None
    best_poly_rmse = None
    print("No valid polynomial degree found (all R^2 were negative or no degrees were evaluated).")

if best_polynomial_degree is not None:
    print(f"\nBest polynomial degree based on test R^2: {best_polynomial_degree}")
    print(f"Corresponding Test R^2: {best_poly_r2:.4f}")
    print(f"Corresponding Test RMSE: {best_poly_rmse:.4f}")

    # Train the final model using the best identified polynomial degree on the entire transformed training set.
    best_polynomial_converter = PolynomialFeatures(degree=best_polynomial_degree, include_bias=False)
    X_train_best_poly = best_polynomial_converter.fit_transform(X_train_numeric)
    X_test_best_poly = best_polynomial_converter.transform(X_test_numeric)

    best_poly_model = LinearRegression()
    best_poly_model.fit(X_train_best_poly, y_train)

    # Evaluate this best polynomial model on the transformed test set (already done above, but re-calculating for clarity)
    y_test_pred_best_poly = best_poly_model.predict(X_test_best_poly)
    final_best_poly_r2 = r2_score(y_test, y_test_pred_best_poly)
    final_best_poly_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred_best_poly))

    print("\nPerformance of the best polynomial model on the test set:")
    print(f"R^2: {final_best_poly_r2:.4f}")
    print(f"RMSE: {final_best_poly_rmse:.4f}")

    # Compare to the best model from part 4 (the combined model)
    print("\nComparison with the best model from Part 4 (Combined Model):")
    print(f"Combined Model (Part 4): R^2 = {best_r2:.4f}, RMSE = {best_rmse:.4f}")
    print(f"Best Polynomial Model: R^2 = {final_best_poly_r2:.4f}, RMSE = {final_best_poly_rmse:.4f}")

    if final_best_poly_r2 > best_r2:
        print("The best polynomial model performs better than the combined model from Part 4.")
    else:
        print("The best polynomial model does not perform better than the combined model from Part 4.")
else:
    print("\nCould not train the final best polynomial model as no valid degree was found.")
```

```python
# Q3.4
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import pandas as pd
from sklearn.metrics import r2_score, mean_squared_error

# Prepare features and target for the flexible model
# We need 'age' and 'transplant' to create interaction terms
X_flex = heart_df[['age', 'transplant']]
y_flex = heart_df['y']

# Split the data into training and testing sets
X_train_flex, X_test_flex, y_train_flex, y_test_flex = train_test_split(X_flex, y_flex, test_size=0.2, random_state=42)

# Apply one-hot encoding to 'transplant' in both train and test sets
X_train_flex_encoded = pd.get_dummies(X_train_flex, columns=['transplant'], drop_first=True)
X_test_flex_encoded = pd.get_dummies(X_test_flex, columns=['transplant'], drop_first=True)

# Ensure consistent columns after encoding (handle potential missing categories in test set)
train_cols_flex = X_train_flex_encoded.columns
test_cols_flex = X_test_flex_encoded.columns

missing_in_test_flex = set(train_cols_flex) - set(test_cols_flex)
for c in missing_in_test_flex:
    X_test_flex_encoded[c] = 0

missing_in_train_flex = set(test_cols_flex) - set(train_cols_flex)
for c in missing_in_train_flex:
    X_train_flex_encoded[c] = 0

# Ensure the order of columns is the same
X_test_flex_encoded = X_test_flex_encoded[train_cols_flex]

# Create polynomial features for 'age' for non-linear effects. Let's try degree 2 for age.
# Also, create interaction terms between polynomial age features and the transplant dummy.
# We need to add age squared and the interaction term age_squared * transplant_treatment
# Let's use PolynomialFeatures on 'age' and then manually create interaction with 'transplant_treatment'
poly = PolynomialFeatures(degree=2, include_bias=False)

# Fit and transform 'age' column
age_train_poly = poly.fit_transform(X_train_flex_encoded[['age']])
age_test_poly = poly.transform(X_test_flex_encoded[['age']])

# Create DataFrames from polynomial age features
age_train_poly_df = pd.DataFrame(age_train_poly, index=X_train_flex_encoded.index, columns=[f'age_poly_{i}' for i in range(age_train_poly.shape[1])])
age_test_poly_df = pd.DataFrame(age_test_poly, index=X_test_flex_encoded.index, columns=[f'age_poly_{i}' for i in range(age_test_poly.shape[1])])


# Combine polynomial age features with the transplant dummy
X_train_flex_final = pd.concat([age_train_poly_df, X_train_flex_encoded[['transplant_treatment']]], axis=1)
X_test_flex_final = pd.concat([age_test_poly_df, X_test_flex_encoded[['transplant_treatment']]], axis=1)

# Create interaction term: age_poly_1 (age squared) * transplant_treatment
# Note: age_poly_0 is just 'age' from degree 1, age_poly_1 is 'age^2' from degree 2
X_train_flex_final['age_poly_1_x_transplant'] = X_train_flex_final['age_poly_1'] * X_train_flex_final['transplant_treatment']
X_test_flex_final['age_poly_1_x_transplant'] = X_test_flex_final['age_poly_1'] * X_test_flex_final['transplant_treatment']

# Add a constant to the final feature sets
X_train_flex_final = sm.add_constant(X_train_flex_final)
X_test_flex_final = sm.add_constant(X_test_flex_final)

# Explicitly convert all columns to float type
X_train_flex_final = X_train_flex_final.astype(float)
X_test_flex_final = X_test_flex_final.astype(float)


# Fit the flexible linear regression model
flexible_model = sm.OLS(y_train_flex, X_train_flex_final).fit()

# Evaluate the flexible model on the test set
y_test_pred_flexible = flexible_model.predict(X_test_flex_final)

# Calculate R^2 and RMSE on the test set
r2_test_flexible = r2_score(y_test_flex, y_test_pred_flexible)
rmse_test_flexible = np.sqrt(mean_squared_error(y_test_flex, y_test_pred_flexible))

print("\nQ3 Part 4:")
print("Flexible Model Performance (Test Set):")
print(f"Test R^2: {r2_test_flexible:.4f}")
print(f"Test RMSE: {rmse_test_flexible:.4f}")
print("\nFlexible Model Summary:")
print(flexible_model.summary())


# Predict survival probability by age for plotting
# Create a range of ages for prediction
age_range = np.linspace(heart_df['age'].min(), heart_df['age'].max(), 100).reshape(-1, 1)

# Create dummy variables for transplant status for the age range
transplant_control_dummy = pd.DataFrame({'transplant_treatment': np.zeros(100)})
transplant_treatment_dummy = pd.DataFrame({'transplant_treatment': np.ones(100)})

# Create polynomial features for the age range
age_range_poly = poly.transform(age_range)
age_range_poly_df = pd.DataFrame(age_range_poly, columns=[f'age_poly_{i}' for i in range(age_range_poly.shape[1])])


# Prepare data for prediction for control group
X_predict_control = pd.concat([age_range_poly_df, transplant_control_dummy], axis=1)
X_predict_control['age_poly_1_x_transplant'] = X_predict_control['age_poly_1'] * X_predict_control['transplant_treatment'] # Interaction term
X_predict_control = sm.add_constant(X_predict_control)
# Ensure columns match the training data (this is important for prediction)
X_predict_control = X_predict_control.reindex(columns=X_train_flex_final.columns, fill_value=0)


# Prepare data for prediction for treatment group
X_predict_treatment = pd.concat([age_range_poly_df, transplant_treatment_dummy], axis=1)
X_predict_treatment['age_poly_1_x_transplant'] = X_predict_treatment['age_poly_1'] * X_predict_treatment['transplant_treatment'] # Interaction term
X_predict_treatment = sm.add_constant(X_predict_treatment)
# Ensure columns match the training data
X_predict_treatment = X_predict_treatment.reindex(columns=X_train_flex_final.columns, fill_value=0)

# Explicitly convert all prediction dataframes to float type
X_predict_control = X_predict_control.astype(float)
X_predict_treatment = X_predict_treatment.astype(float)


# Predict survival probability for both groups
predicted_survival_control = flexible_model.predict(X_predict_control)
predicted_survival_treatment = flexible_model.predict(X_predict_treatment)

# Clip predictions to be within [0, 1] as survival probability must be between 0 and 1
predicted_survival_control = np.clip(predicted_survival_control, 0, 1)
predicted_survival_treatment = np.clip(predicted_survival_treatment, 0, 1)


# Plot the predicted survival probability by age, conditional on transplant status
plt.figure(figsize=(10, 6))
plt.plot(age_range, predicted_survival_control, label='Control (No Transplant)', color='blue')
plt.plot(age_range, predicted_survival_treatment, label='Treatment (Transplant)', color='red')
plt.xlabel("Age")
plt.ylabel("Predicted Survival Probability")
plt.title("Predicted Survival Probability by Age and Transplant Status")
plt.legend()
plt.grid(True)
plt.show()
```

**Plot description**

The plot shows the estimated relationship between age and predicted survival probability for both the control and treatment groups.For the control group (blue line), the predicted survival probability generally decreases with increasing age, which is expected. For the treatment group (red line), the predicted survival probability also tends to decrease with age. The red line (treatment group) is generally above the blue line (control group) across most of the age range, suggesting a higher predicted survival probability for transplant recipients, especially in their younger to middle ages.The difference in predicted survival probability between the two groups appears to change with age, indicated by the lines not being parallel. This is due to the interaction term included in the model. At younger ages, the predicted survival benefit of a transplant seems larger. At older ages, the predicted survival benefit appears to diminish, and the lines might even converge or cross, suggesting that a transplant might not offer a survival advantage, or could even be detrimental, for very old patients. The non-linear shape of the curves (due to the polynomial age term) suggests that the rate of decline in survival probability with age is not constant.

**Concerns about using these kinds of models to select organ transplant recipients:**

1. **Data Bias:**
- The model is trained on historical data, which may reflect existing biases in who received transplants in the past (e.g., based on socioeconomic status, race, or other factors not explicitly in the model but correlated with included features). Deploying such a model could perpetuate or even amplify these historical biases.

- The outcome variable ('y': survival for three years) is a simplified measure. Long-term survival, quality of life, and other factors are also crucial but not captured by this simple binary outcome.

2. **Model Limitations:**
- A linear regression model, even with polynomial and interaction terms, might not fully capture the complex non-linear relationships and interactions in survival outcomes. Other models (e.g., survival analysis models like Cox proportional hazards or more complex machine learning models) might be more appropriate.
- The R-squared value on the test set ({:.4f}) is relatively low, indicating that a significant portion of the variability in survival is not explained by the model. This means the model's predictions have a considerable degree of uncertainty.
- The model might not account for all relevant factors influencing survival (e.g., severity of heart disease, comorbidities, lifestyle factors, access to post-transplant care). Omitting important variables can lead to biased estimates and unfair decisions.
- The model provides a predicted probability, but the decision threshold for who receives a transplant is not clear and has significant ethical implications.

3. **Ethical and Fairness Concerns:**
- Using an algorithm for transplant selection raises serious ethical questions about fairness, equity, and justice. Who defines the objective function the algorithm optimizes (e.g., maximizing the number of survivors, maximizing life-years gained)?
- There is a risk of disparate impact on certain demographic groups if the model is biased or if the factors it relies on are correlated with protected characteristics.
- The lack of transparency in complex models can make it difficult to understand why a particular individual was denied a transplant, raising issues of accountability and trust.
- Reducing a complex medical and ethical decision to a single score or probability can dehumanize the process.

4. **Deployment Challenges:**
- How would the model be integrated into the existing transplant allocation system? Would it replace human decision-making or assist it?
- How would the model be maintained and updated? Medical knowledge and patient demographics change over time.
- How would potential errors or biases in the model be identified and addressed in practice?
- There are legal and regulatory challenges associated with using algorithmic decision-making in healthcare, particularly for high-stakes decisions like organ allocation.
- Patient and public acceptance of an algorithm making such critical decisions is uncertain.
